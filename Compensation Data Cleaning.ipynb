```python
#Dropping Duplicates
df=df.drop_duplicates()
```

```python
#Renaming Columns
df= df.rename(columns= { "What is your Current Role?":"Role",
                     "If other, please indicate your role here":"Other_Role",
                     "What is your Level?":"Level",
                    "How many years of experience do you have in the data Field?":"Years_of_Experience",
                    "What industry is  your Employer? eg Fintech, Utilities, HR, Gaming, Health etc.":"Industry",
                    "What is your gender":"Gender",
                    "What is your main Tech stack?":"Stack",
                    "If other, please indicate your tool stack here":"Other_Stack",
                    "What is your monthly Gross Salary in Kes per month?":"Salary",
                    "Any other benefits?":"Other_Benefit",
                    "What is your work setup?":"Work_Setup",
                    "Employer Type":"Employer_Type"
        })
df
```

```python
df.isnull().any()
```

```python
df["Industry"]=df["Industry"].str.strip()
df["Industry"]
```

```python
#Cleaning "Industry" Column- Correcting spelling and grouping the data
df["Industry"]=df["Industry"].replace("Nil","None")
df["Industry"]=df["Industry"].replace("machine learning","Machine Learning")
df["Industry"]=df["Industry"].replace("Professional Services (Accounting)","Accounting")
df["Industry"]=df["Industry"].replace("Fintech","FinTech")
df["Industry"]=df["Industry"].replace("Fast Moving Consumer Goods","FMCG")
df["Industry"]=df["Industry"].replace("Development - INGO","NGO")
df["Industry"]=df["Industry"].replace("Finance - Pensions","Finance")
df["Industry"]=df["Industry"].replace("Fmcg","FMCG")
df["Industry"]=df["Industry"].replace("Healtg","Health")
df["Industry"]=df["Industry"].replace("Analytics->I work for a company in Europe not Kenya","Analytics")
df["Industry"]=df["Industry"].replace("Still searching","None")
df["Industry"]=df["Industry"].replace("health","Health")
df["Industry"]=df["Industry"].replace("Finetech","FinTech")
df["Industry"]=df["Industry"].replace("Manufacturing (FMCG)","FMCG")
df["Industry"]=df["Industry"].replace("Azubi Africa","Analytics")
df["Industry"]=df["Industry"].replace("SAAS(Foodtech)","SAAS")
df["Industry"]=df["Industry"].replace("Telco","Telecommunications")
df["Industry"]=df["Industry"].replace("Government- taxation","Government")
df["Industry"]=df["Industry"].replace("Technology and Telecommunications","Telecommunications")
df["Industry"]=df["Industry"].replace("Bank","Banking")
df["Industry"]=df["Industry"].replace("Sustainable energy","Energy")
df["Industry"]=df["Industry"].replace("People and Technology","Technology")
df["Industry"]=df["Industry"].replace("Market research","Research")
df["Industry"]=df["Industry"].replace("Consumer tech and media","Media")
df["Industry"]=df["Industry"].replace("Wholesale and Retail","Retail")
df["Industry"]=df["Industry"].replace("Production and Retail","Retail")
df["Industry"]=df["Industry"].replace("FMCG consulting","FMCG")
df["Industry"]
```

```python
df["Level"]=df["Level"].str.strip()
df["Level"]= df["Level"].replace("Mid-Level  eg Data Analyst","Mid-Level")
df["Level"]= df["Level"].replace("Junior  eg Junior Data Analyst","Junior")
df["Level"]= df["Level"].replace("Manager eg Manager of Analytics","Manager")
df["Level"]= df["Level"].replace("Senior Level  eg Senior Data Analyst","Senior")
df["Level"]= df["Level"].replace("Lead eg Lead Analyst","Lead")
df["Level"]
```

```python
#Salary column cleaning- Changing different currency format and converting them to standard currency i.e Ksh
df["Salary"]=df["Salary"].replace("00", "0")
df["Salary"]=df["Salary"].replace("500$", "65000")
df["Salary"]=df["Salary"].replace("15,000/-", "15000")
df["Salary"]=df["Salary"].replace("NaN", "0")
df["Salary"]=df["Salary"].replace("112,000KES", "112000")
df["Salary"]=df["Salary"].replace("920 EUROS", "130000")
df["Salary"]=df["Salary"].replace("Unemployed", "0")
df["Salary"]=df["Salary"].replace("Nill", "0")
df["Salary"]=df["Salary"].replace("25", "25000")


df["Salary"]
```

```python
df["Work_Setup"]= df["Work_Setup"].replace("Hybrid-I choose my days to go the office", "Hybrid-Choose Office Days")
df["Work_Setup"]= df["Work_Setup"].replace("Hybrid- I need to be in the office specific days in a week", "Hybrid-Specific Office Days")
df["Work_Setup"]
```

```python
#Droppind "Timestamp" column
df=df.drop(columns="Timestamp")
df
```

```python
#Replacing null values with 0 to be able to change data type
df["Salary"]= df["Salary"].fillna(0)
```
